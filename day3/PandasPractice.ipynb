```python
import pandas as pd
import numpy as np
```

```python
# Load the data
df = pd.read_csv('../data/frog_tongue_adhesion.csv', comment='#')

# Take a look
df.head()
```

```python
#practice 1 a
df.loc[np.abs(df["adhesive strength (Pa)"])>= 2000,
        ['adhesive strength (Pa)','impact time (ms)']
       ]
```

```python
#Practice 1.b
df.loc[df["ID"] == 'II',
        ['impact force (mN)','adhesive force (mN)']
       ].reset_index()
```

```python
#Practice 1.c

# Extract the adhesive force and the time the frog pulls on the target 
# for juvenile frogs (Frogs III and IV)

df.loc[
    ((df['ID']== 'III')|(df['ID'] =='IV')),
    :
]

#eq to 
df.loc[
    df['ID'].isin(['III', 'IV']),
      :
]
```

```python
# Practice 2.a

grouped = df.groupby('ID')
grouped['impact force (mN)'].std().reset_index()
```

```python
#Practice 2.b

def coeff_of_var(data):
    """Compute coefficient of variation from an array of data."""
    return np.std(data) / np.abs(np.mean(data))


grouped = df.groupby('ID')
grouped['impact force (mN)','adhesive force (mN)'].agg(coeff_of_var).reset_index()
```

```python
#Practice 2.c

grouped = df.groupby('ID')['impact force (mN)','adhesive force (mN)']

df_analysed = grouped.agg([np.mean, np.median, np.std, coeff_of_var]).reset_index()

df_analysed
```

```python
# Melt the DataFrame to make it tidy
#takes every entry and puts it in a column name 'value' by default
#id_vars days which column is not made of values, not to be put in the value column
# give names of columns (come from previous headings) in order
#headings become columns to be used for boolean indexing
pd.melt(df_analysed, var_name=['quantity', 'statistic'], id_vars='ID')
```
