```python
def coeff_of_var(data):
    """Compute coefficient of variation from an array of data."""
    return np.std(data) / np.mean(data)
```

```python
grouped = df.groupby(['gender', 'insomnia'])

grouped['percent correct'].agg(coeff_of_var).reset_index()
```

```python
import pandas as pd

# Our main plotting package (must have explicit import of submodules)
import bokeh.io
import bokeh.plotting

# Enable viewing Bokeh plots in the notebook
bokeh.io.output_notebook()
```

```python
p = bokeh.plotting.figure(
    height=300,
    width=400,
   # background_fill_color='purple',
    #background_fill_alpha=0.3,
    x_axis_label= 'confidence when correct',
    y_axis_label='confidence when incorrect'
)
```

```python
p.circle(
    source=df,
    x='confidence when correct',
    y='confidence when incorrect',
    size=5,
    color='dodgerblue',
    
)
```

```python
bokeh.io.show(p)
```

```python
x='confidence when correct'
y='confidence when incorrect'

p =bokeh.plotting.figure(
    height=300,
    width=400,
    x_axis_label=x,
    y_axis_label=y,
    tooltips = [
        ('p-number', '@{participant number}'),
        ('gender', '@gender'),
        ('correctness', '@{percent correct}'),
    ]
)

p.circle(
    source=df.loc[~df['insomnia'],:],
    x=x,
    y=y,
    legend='normal sleepers',

)

p.circle(
    source=df.loc[df['insomnia'],:],
    x=x,
    y=y,
    color='orange',
    legend='insomiacs',

)
p.legend.location = 'top_left'
p.legend.click_policy = 'hide'
bokeh.io.show(p)
```

```python
bokeh.io.save(p,)
```

```python
import numpy as np
import pandas as pd

import bokeh_catplot

import bokeh.plotting
import bokeh.io

bokeh.io.output_notebook()
```

```python
df = pd.read_csv('../data/frog_tongue_adhesion.csv', comment='#')
df.head()
```

```python
p = bokeh_catplot.box(
    data=df,
    cats='ID',
    val='impact force (mN)',
)

bokeh.io.show(p)
```
