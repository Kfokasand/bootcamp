```python
import pandas as pd

import bokeh.io
import bokeh.plotting

bokeh.io.output_notebook()
```

```python
# Load in the data
df = pd.read_csv('../data/collins_switch.csv', comment='#')
df
```

```python

x = '[IPTG] (mM)'
y = 'normalized GFP expression (a.u.)'

p = bokeh.plotting.figure(
    height=300,
    width=400,
    x_axis_type='log',
    x_axis_label=x,
    y_axis_label=y,
)

p.circle(
    source=df,
    x=x,
    y=y,

)

bokeh.io.show(p)
```

```python
# Add error bars to the DataFrame
df['error_low'] = df['normalized GFP expression (a.u.)'] - 1.96*df['sem']
df['error_high'] = df['normalized GFP expression (a.u.)'] + 1.96*df['sem']

# Add error bars
p.segment(
    source=df,
    x0=x,
    y0='error_low',
    x1=x,
    y1='error_high',
    line_width=2
)

bokeh.io.show(p)
```
