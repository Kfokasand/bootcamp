```python
# add inscomniac column

df["insomnia"] = df["sci"] <= 16

df["bleb"] = "Nan"
del df["bleb"]
df.head()
```

```python
#ranking of subjects according to correct percentage within gender and insomnia groups

df.groupby(['gender','insomnia'])['percent correct'].rank()

df.head()
```

```python
def coeff_of_var(data):
    """Compute coefficient of variation from an array of data."""
    return np.std(data) / np.mean(data)
```

```python
grouped = df.groupby(['gender', 'insomnia'])

grouped['percent correct'].agg(coeff_of_var).reset_index()
```
